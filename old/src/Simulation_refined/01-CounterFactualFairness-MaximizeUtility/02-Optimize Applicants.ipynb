```python
from ortools.linear_solver import pywraplp
```

```python
solver = pywraplp.Solver.CreateSolver('GLOP')
```

```python
df_ = df.head(n=100)
```

```python
applicants = []

# Objective: minimize the sum of (price-normalized) foods.
objective = solver.Objective()
for ix, row in df_.iterrows():
    applicants.append(solver.NumVar(0.0, 1.0, str(ix)))
    objective.SetCoefficient(applicants[ix], row['ml_outcomes'])
objective.SetMaximization()
```

```python
constraints = [0] * 1
```

```python
K = 10
k_admits = solver.Constraint(0, K)
```

```python
for ix, row in df_.iterrows():
    k_admits.SetCoefficient(applicants[ix], 1.0)
```

```python
status = solver.Solve()
```

```python
status
```

```python
solver.OPTIMAL
```

```python
row = []
admit = []

for i in applicants:
    row.append(int(str(i)))
    admit.append(i.solution_value())

df_decisions = pd.DataFrame({'row_id':row,'decision':admit})
```
